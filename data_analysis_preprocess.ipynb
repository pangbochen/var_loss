## set label for classification

```python
high_quantile = 0.80
low_quantile = 0.20
```

```python
data_df['return'].quantile(high_quantile)
```

```python
data_df['return'].quantile(low_quantile)
```

```python
def set_label_3(x, high, low):
    if x > high:
        return 0 # 多
    elif x < low:
        return 2 # 空
    else:
        return 1 # 波动
```

```python
high = data_df['return'].quantile(high_quantile)
low = data_df['return'].quantile(low_quantile)
data_df['label_3'] = data_df['return'].apply(lambda x: set_label_3(x, high, low))
```

```python
data_df.head()
```

```python
def set_label_2(x):
    if x > 0:
        return 0 # 多
    else:
        return 1 # 空
```

```python
data_df['label_2'] = data_df['return'].apply(lambda x: set_label_2(x))
```
